```python
%pylab inline
import cPickle as pickle
import numpy as np
import itertools
```

```python
with open("data/ratings.pickle", "rb") as f:
    ratings = pickle.load(f)
ratings_shuffled = np.random.permutation(ratings)
```

```python
class SimpleDistributionModel(object):
    def __init__(self, training_set):
        rating_totals = np.ones(5)
        for biz, pid, rating in training_set:
            rating_totals[int(float(rating))-1] += 1
        self.rating_distribution = rating_totals / rating_totals.sum()
    def predict(self, business_id, person_id):
        return self.rating_distribution
```

```python
def prediction_from_distribution(distribution):
    return np.argmax(distribution)+1
```

```python
# assign a training set and a cross-validation set to see how well we're doing
# 17167 is 1/4 of the total ratings
training_set = ratings_shuffled[:-17167]
training_set_ratings = np.array([int(float(t)) for t in training_set[:,2]])
cross_validation_set = ratings_shuffled[-17167:]
cross_validation_set_ratings = np.array([int(float(t)) for t in cross_validation_set[:,2]])
```

```python
cv_error = []
t_error = []
for training_set_size in xrange(1, 100): #len(training_set)+1):
    # train it
    model = SimpleDistributionModel(training_set[:training_set_size])
    # test it
    total_correct = 0
    for actual, test in itertools.izip(training_set_ratings[:training_set_size], training_set[:training_set_size]):
        total_correct += actual == prediction_from_distribution(model.predict(test[0], test[1]))
    t_error.append(1.0 - float(total_correct) / training_set_size)
    # test on the whole cross-validation set
    total_correct = 0
    for actual, test in itertools.izip(cross_validation_set_ratings, cross_validation_set):
        total_correct += actual == prediction_from_distribution(model.predict(test[0], test[1]))
    cv_error.append(1.0 - float(total_correct) / len(cross_validation_set))
```

```python
plot(range(1, len(t_error)+1), t_error, label="training set")
plot(range(1, len(cv_error)+1), cv_error, label="cross-validation set")
title("Learning curves for SimpleDistributionModel")
ylabel("error (1.0=max)")
xlabel("training set size")
legend()
```

For this very simple model, more data will not help, since error for both the training set and the cross-validation have high error (and about the same). This means the model has high bias.

```python
def doexperiment(modelclass, iterations, training_set, cross_validation_set):
    if iterations > len(training_set):
        iterations = len(training_set)
    cv_error = []
    t_error = []
    for training_set_size in xrange(1, iterations): #len(training_set)+1):
        # train it
        model = modelclass(training_set[:training_set_size])
        # test it
        total_correct = 0
        for test in training_set[:training_set_size]:
            actual = int(float(test[2]))
            total_correct += actual == prediction_from_distribution(model.predict(test[0], test[1]))
        t_error.append(1.0 - float(total_correct) / training_set_size)
        # test on the whole cross-validation set
        total_correct = 0
        for test in cross_validation_set:
            actual = int(float(test[2]))
            total_correct += actual == prediction_from_distribution(model.predict(test[0], test[1]))
        cv_error.append(1.0 - float(total_correct) / len(cross_validation_set))
    return cv_error, t_error
```

```python
cv_error, t_error = doexperiment(SimpleDistributionModel, 100, training_set, cross_validation_set)
```

```python
def plotlearningcurve(cv_error, t_error, modelname):
    plot(range(1, len(t_error)+1), t_error, label="training set")
    plot(range(1, len(cv_error)+1), cv_error, label="cross-validation set")
    title("Learning curves for {model}".format(model=modelname))
    ylabel("error (1.0=max)")
    xlabel("training set size")
    legend()
```

```python
plotlearningcurve(cv_error, t_error, "simple mode-based model")
```

```python
class PerBusinessSimpleDistributionModel(object):
    def __init__(self, training_set):
        rating_totals = np.ones(5)
        businesses = {}
        for biz, pid, rating in training_set:
            rating_totals[int(float(rating))-1] += 1
            if biz not in businesses:
                businesses[biz] = np.ones(5)
            businesses[biz][int(float(rating))-1] += 1
        self.businesses = businesses
        self.rating_distribution = rating_totals / rating_totals.sum()
    def predict(self, business_id, person_id):
        if business_id in self.businesses:
            return self.businesses[business_id]
        return self.rating_distribution
```

```python
# I'm having to do many more trials to get the test error up.
# How many businesses have multiple ratings?
cv_error, t_error = doexperiment(PerBusinessSimpleDistributionModel, 1000, training_set, cross_validation_set)
```

```python
plotlearningcurve(cv_error, t_error, "simple per-business mode-based model")
```

Well, going per-business doesn't seem to help much at all, does it? This model has very high variance, though. High variance makes sense, since many reviews in the test set will be the only one for that business.

```python
class PerReviewerSimpleDistributionModel(object):
    def __init__(self, training_set):
        rating_totals = np.ones(5)
        reviewers = {}
        for biz, pid, rating in training_set:
            rating_totals[int(float(rating))-1] += 1
            if pid not in reviewers:
                reviewers[pid] = np.ones(5)
            reviewers[pid][int(float(rating))-1] += 1
        self.reviewers = reviewers
        self.rating_distribution = rating_totals / rating_totals.sum()
    def predict(self, business_id, person_id):
        if person_id in self.reviewers:
            return self.reviewers[person_id]
        return self.rating_distribution
```

```python
cv_error, t_error = doexperiment(PerReviewerSimpleDistributionModel, 2500, training_set, cross_validation_set)
```

```python
plotlearningcurve(cv_error, t_error, "simple per-reviewer mode-based model")
```

Clearly, this means there are fewer friends than businesses in these reviews, but it seems that the distributions between Yelp reviewers probably aren't all that different.
